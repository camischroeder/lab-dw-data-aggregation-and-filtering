# Lab | Data Aggregation and Filtering

In this challenge, we will continue to work with customer data from an insurance company. We will use the dataset called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd

mkt_cst_analysis = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")

mkt_cst_analysis["Customer Lifetime Value"] = mkt_cst_analysis["Customer Lifetime Value"].round(1)
mkt_cst_analysis["Total Claim Amount"] = mkt_cst_analysis["Total Claim Amount"].round(1)
mean_income = mkt_cst_analysis["Income"].mean()
mkt_cst_analysis["Income"] = mkt_cst_analysis["Income"].replace(0, mean_income)
mkt_cst_analysis["Income"] = mkt_cst_analysis["Income"].round(1)
mkt_cst_analysis = mkt_cst_analysis.drop_duplicates()
mkt_cst_analysis = mkt_cst_analysis.dropna()
mkt_cst_analysis = mkt_cst_analysis.reset_index(drop = True)
mkt_cst_analysis = mkt_cst_analysis.drop("Unnamed: 0", axis = 1)

mkt_cst_analysis
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
total_claim_vs_positive_response = mkt_cst_analysis
total_claim_vs_positive_response = total_claim_vs_positive_response[total_claim_vs_positive_response["Total Claim Amount"] > 1000]
total_claim_vs_positive_response = total_claim_vs_positive_response[total_claim_vs_positive_response["Response"] == "Yes"]

total_claim_vs_positive_response
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
columns2 = ["Gender", "Policy", "Total Claim Amount", "Response"]
claim_policy_gender_resp_yes = mkt_cst_analysis[columns2]
claim_policy_gender_resp_yes = claim_policy_gender_resp_yes[claim_policy_gender_resp_yes["Response"] == "Yes"]
claim_policy_gender_resp_yes = claim_policy_gender_resp_yes.sort_values(by = "Policy")
claim_policy_gender_resp_yes = claim_policy_gender_resp_yes.reset_index(drop = True)

claim_policy_gender_resp_yes
```

```python
claim_policy_gender_resp_yes = claim_policy_gender_resp_yes.groupby(['Gender', 'Policy'])['Total Claim Amount'].mean().round(1).reset_index()
claim_policy_gender_resp_yes = claim_policy_gender_resp_yes.sort_values(by = "Policy")

claim_policy_gender_resp_yes
```

'''
- on all personal policies and corporate L3, women have a smaller average claim amount than men
- the most diverging average claim amount result between genders in for the "Special L2" policy - 120% higher for women.
'''

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
bystate_higher_five = mkt_cst_analysis.groupby("State")["Customer"].count().reset_index()
bystate_higher_five_final = bystate_higher_five[bystate_higher_five["Customer"] > 500]

bystate_higher_five_final
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
max_min_median = mkt_cst_analysis.groupby(["Education", "Gender"])['Customer Lifetime Value'].agg(['max', 'min', 'median']).reset_index()
max_min_median.columns = ["Education", "Gender", "Max CLV", "Min CLV", "Median CLV"]

max_min_median
```
